**2 Data retrieve**


```python
# Import libraries and modules
# Functions from pandas_datareader.data extract data from Yahoo Finance into a DataFrame

# import pandas_datareader.data as web
# import datetime
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
# Extracting data from UK stock indexed
# from Yahoo Finance

! pip install yfinance
import yfinance as yf

# FTSE 100 stock index (UK)

# The first line of capital letters is the tickers for each stock index
# As previously stated, I have set the time between 31 December 2019 and 15 February 2022
# so the start is 2019-12-31, end is 2022-02-15

FTSE100 = yf.download('^FTSE',
start='2019-12-31',
end='2022-02-15',
progress=False)

# Check that the table header is correct
FTSE100.head()
```

```python
# FTSE 350 stock index (UK)
FTSE350 = yf.download('^FTLC',
start='2019-12-31',
end='2022-02-15',
progress=False)

# Check that the table header is correct
FTSE350.head()
```

```python
# FTSE All-Share stock index (UK)
FTSE_all_share = yf.download('^FTAS',
start='2019-12-31',
end='2022-02-15',
progress=False)

# Check that the table header is correct
FTSE_all_share.head()
```

```python
# Extracting data from US stock indexed
# from Yahoo Finance

# NASDAQ stock index (US)
NASDAQ = yf.download('^IXIC',
start='2019-12-31',
end='2022-02-15',
progress=False)

# Check that the table header is correct
NASDAQ.head()
```

```python
# SP500 stock index (US)
SP500 = yf.download('^GSPC',
start='2019-12-31',
end='2022-02-15',
progress=False)

# Check that the table header is correct
SP500.head()
```

```python
# Dow Jones stock index (US)
Dow_jones = yf.download('^DJI',
start='2019-12-31',
end='2022-02-15',
progress=False)

# Check that the table header is correct
Dow_jones.head()
```

```python
# Convert to csv files and save 
FTSE100.to_csv('FTSE100_stock.csv') 
FTSE350.to_csv('FTSE350_stock.csv') 
FTSE_all_share.to_csv('FTSE_all_share_stock.csv')
NASDAQ.to_csv('NASDAQ_stock.csv') 
SP500.to_csv('SP500_stock.csv') 
Dow_jones.to_csv('Dow_jones_stock.csv')
```

**4.2 Reasons of the trend**

To find out why stock indexes are moving this way, I will locate the point in time when the stock went up or down. Then find out what happened at that point in time, and then that thing may be the reason why the stock moved higher or lower.

I will analyse the highs and lows of stock market and trade volume trends. 

Data would be normalized prior to analysis. 



```python
# UK stock price normalization 


# make a copy 
# as the nomalization part may change the original data 
FTSE100_norm=FTSE100.copy()
FTSE350_norm=SP500.copy()
FTSE_all_share_norm=FTSE_all_share.copy()


from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()

# Normalize Open and Volume data in UK stock indexes
FTSE100_norm[['Open','Volume']]=mms.fit_transform(FTSE100_norm[['Open','Volume']])
FTSE350_norm[['Open','Volume']]=mms.fit_transform(FTSE350_norm[['Open','Volume']])
FTSE_all_share_norm[['Open','Volume']]=mms.fit_transform(FTSE_all_share[['Open','Volume']])

# Plot UK stock indexes price normailzation
FTSE100_norm['Open'].plot(label = 'FTSE 100 open price', figsize = (30,15))
FTSE350_norm['Open'].plot(label = 'FTSE 350 open price', figsize = (30,15))
FTSE_all_share_norm['Open'].plot(label = 'FTSE All-Share open price', figsize = (30,15))

plt.title("UK Market Price Normalization")
plt.ylabel('Stock Price')
plt.legend()
```

```python
# Look at the time of lowest point in FTSE 100  

# Return position of the lowest value
lowest_FTSE100 = FTSE100.iloc[FTSE100['Open'].argmin()]
lowest_FTSE100 = pd.DataFrame(lowest_FTSE100)
display(lowest_FTSE100)
```

```python
# Look at the time of lowest point in FTSE 350   
lowest_FTSE350 = FTSE350.iloc[FTSE350['Open'].argmin()]
lowest_FTSE350 = pd.DataFrame(lowest_FTSE350)
display(lowest_FTSE350)
```
